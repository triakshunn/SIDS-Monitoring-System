```python
import pyrebase
import time
config = {
  "apiKey": "**************************",
  "authDomain": "**********************",
  "databaseURL": "************************",
  "storageBucket": "******************************"
}

firebase = pyrebase.initialize_app(config)
```

```python
age=int(input("Enter the age of the child(in months)"))
race=int(input("Enter your race[1 for Central Americans and South Americans,2 for Asian/Pacific Islanders,3 for Mexicans,4 for Puerto Ricans,5 for Whites,6 for African Americans,7 for Native American,8 for others"))
gender=int(input("Enter 1 for Boy,2 for Girl,3 for others"))
parent_beh=int(input("Do you smoke/consume alcohol? Enter 1 if yes and 0 if no"))
baby_conditions=int(input("Did the baby had any past of any health complications? Enter 1 if yes"))
```

```python
if (age<4) and (age>2):
  age=1
elif age<2 or (age>4 and age<12):
  age=0.5
else:
  age=0
if race==1:
  race=0.03
elif race==2 or race==3:
  race=0.04
elif race==4 or race==5:
  race=0.08
elif race==6 or race==7:
  race=0.15
else:
  race=0.08
if gender==1:
  gender=1;
else:
  gender=0.5
```

```python
db=firebase.database()
```

```python
def SidsProb(angle,child_age,race,gender,parent_beh,room_temp,baby_conditions):
  if child_age==1:
    return (0.55*angle + race + 0.1*gender+ 0.1*parent_beh+0.05*room_temp+0.05*baby_conditions)
  elif child_age==0.5:
    return(0.35*angle + race + 0.1*gender+ 0.1*parent_beh+0.05*room_temp+0.05*baby_conditions)
  else:
    return(0.15*angle + race + 0.05*gender+ 0.05*parent_beh+0.05*room_temp+0.05*baby_conditions)
prob=SidsProb(1,1,0.1,1,1,1,1)
print(prob)
```

```python
heart_beat_arr=[0,0,0,0,0]
```

```python
while(1):
  users_angle=db.child("test").child("Y_angle").get()
  users_temp=db.child("test").child("Temp").get()
  users_heartbeat=db.child("test").child("Heart").get()
  last_angle=list(users_angle.val().keys())[-1]
  last_temp=list(users_temp.val().keys())[-1]
  last_heartbeat=list(users_heartbeat.val().keys())[-1]
  y_axis=(users_angle.val()[last_angle])
  temp=(users_temp.val()[last_temp])
  heartbeat=(users_heartbeat.val()[last_heartbeat])
  heart_beat_arr.append(heartbeat)
  heart_beat_sum=sum(heart_beat_arr[-5:])
  if temp<15 or temp>40:
      temp_tem=1
  else:
      temp_tem=0
  if abs(y_axis)>40:
      angle=1
  else:
      angle=0
  if heart_beat_sum==0:
    print(" NO HEART BEAT DETECTED; HIGH SIDS RISK  ")
    print("-----------------------------------------")
  else:
    sids_prob=SidsProb(angle,age,race,gender,parent_beh,temp_tem,baby_conditions)
    if sids_prob>=.55:
       print(" HIGH SIDS RISK ")
       print("SIDS Probability: "+str(sids_prob)+" Rotation Angle: "+str(y_axis)+" Room Temperature: "+str(temp))
       print("-----------------------------------------")
    else:
       print("SIDS Probability: "+str(sids_prob)+" Rotation Angle: "+str(y_axis)+" Room Temperature: "+str(temp))
       print("-----------------------------------------")
  time.sleep(3)
```
